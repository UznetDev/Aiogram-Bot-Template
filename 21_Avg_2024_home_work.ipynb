## Harbir qitada monarxiya soni va respublikalar nechta

```python
%%sql SELECT
    s.continent,
    SUM(CASE WHEN m.monarch IS NOT NULL THEN 1 ELSE 0 END) as num_monarchies,
    SUM(CASE WHEN p.president IS NOT NULL THEN 1 ELSE 0 END) as num_republics
FROM states AS s
LEFT JOIN monarchs AS m ON s.country = m.country
LEFT JOIN presidents AS p ON s.country = p.country
GROUP BY s.continent;
```

## Har bir yil uchun prime ministirlar soni

```python
%%sql SELECT
    SUBSTR(pm_start, 1, 4) AS year,
    COUNT(*) AS num_prime_ministers
FROM prime_minister_terms
WHERE pm_start IS NOT NULL
GROUP BY year
ORDER BY year;
```

```python
%%sql SELECT
    country,
    monarch AS leader,
    'monarch' AS leader_type
FROM monarchs
UNION ALL
SELECT
    country,
    president AS leader,
    'president' AS leader_type
FROM presidents
UNION ALL
SELECT
    country,
    prime_minister AS leader,
    'prime minister' AS leader_type
FROM prime_ministers
ORDER BY country, leader_type;
```

```python
%%sql SELECT
    s.continent,
    COALESCE(m.num_monarchs, 0) AS num_monarchs,
    COALESCE(p.num_presidents, 0) AS num_presidents,
    COALESCE(pm.num_prime_ministers, 0) AS num_prime_ministers
FROM states AS s
LEFT JOIN (SELECT continent,
                  COUNT(*) AS num_monarchs
           FROM monarchs
           GROUP BY continent) AS m
ON s.continent = m.continent
LEFT JOIN (SELECT continent,
                  COUNT(*) AS num_presidents
           FROM presidents
           GROUP BY continent) AS p
ON s.continent = p.continent
LEFT JOIN (SELECT continent,
                  COUNT(*) AS num_prime_ministers
           FROM prime_ministers
           GROUP BY continent) AS pm
ON s.continent = pm.continent
GROUP BY s.continent;
```
